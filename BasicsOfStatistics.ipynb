# Basics Of Statistics

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATASET_SIZE = 1000

x = np.linspace(-1.0, 1.0, DATASET_SIZE)
```

```python
print(x.shape)
print(x.ndim)
print(len(x))
```

## Generate Synthetic Toy Data
$\Large y=ax+b$

```python
y = 3 * x + 5

print(y.shape)
print(y.ndim)
print(len(y))
```

```python
plt.plot(x, y)
plt.title('Toy Synthetic Data')
plt.xlabel('X')
plt.ylabel('Y')
```

## Add Noise

```python
noise = np.random.randn(DATASET_SIZE)

_ = plt.hist(noise, bins=30)
plt.title('Distribution of Gaussian (Normal) Noise')
```

```python
y = y + 0.1*np.random.randn(len(y)) 
```

```python
plt.plot(x, y)
plt.title('Toy Synthetic Data with Noise')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid('on')
```

## Move Center of Data

```python
y = y + 10

plt.plot(x, y)
plt.title('Toy Synthetic Data with Noise SHIFTED')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid('on')
```

```python
y = y - 50

plt.plot(x, y)
plt.title('Toy Synthetic Data with Noise SHIFTED')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid('on')
```

## Mean
$\Large \mu = \frac{1}{N} \sum_{i=1}^{N} x_i$

Where:
- $\mu$ is the mean of the data
- $N$ is the total number of data samples

```python
mean_1 = np.sum(y) / len(y)
mean_2 = np.mean(y)
mean_3 = y.mean()

print(f'Mean manual: {mean_1:.4f}')
print(f'Mean with np.mean: {mean_2:.4f}')
print(f'Mean with y.mean(): {mean_3:.4f}')
```

## Zero Mean

```python
y = y - y.mean()

plt.plot(x, y)
plt.title('Toy Synthetic Data with Noise Zero Mean')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid('on')
```

## Variance ($\sigma^2$), Standart Deviation ($\sigma$)
$\Large \sigma^2 = \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2$

$\Large \sigma = \sqrt{\sigma^2}$

Where:
- $\mu$: is the mean of the data
- $N$ is the total number of data samples

```python
def calculate_variance(array):
    zero_mean = array - array.mean()
    sigma2 = np.mean(zero_mean ** 2)
    return sigma2
```

```python
variance_1 = y.var()
variance_2 = calculate_variance(y)

print(f'Variance y.var(): {variance_1:.4f}')
print(f'Variance manual: {variance_2:.4f}')
```

```python
std_1 = y.std()
std_2 = np.sqrt(variance_2)

print(f'Standart deviation y.std(): {std_1:.4f}')
print(f'Standart deviation manual: {std_2:.4f}')
```

## Variance for Different Data

```python
freq = 1
DATASET_SIZE = 1000

x = np.linspace(-1.0, 1.0, DATASET_SIZE)

y_2 = np.sin(2*np.pi*freq*x) + 0.1*np.random.randn(len(x))

print(f'Std: {y_2.std():.4f}')
print(f'Var: {y_2.var():.4f}')


plt.plot(x, y_2)
plt.title('Sine Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid('on')
```

## Outlier

```python
y = np.random.randn(len(x))

print(f'Std: {y.std():.4f}')
print(f'Var: {y.var():.4f}')

plt.plot(x, y)
plt.title('Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid('on')
```

```python
SPIKE_AMOUNT = 10
y[100:120] = y[100:120] + SPIKE_AMOUNT

print(f'Std: {y.std():.4f}')
print(f'Var: {y.var():.4f}')

plt.plot(x, y)
plt.title('Data with Outliers (Spike)')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid('on')
```

## Percentiles

```python
percentiles = np.percentile(y, [25, 50, 75])

print('25th percentile (Q1):', percentiles[0])
print('50th percentile (Median):', percentiles[1])
print('75th percentile (Q3):', percentiles[2])
```

```python
percentiles = np.percentile(y, [95, 97, 99])

print('95th percentile:', percentiles[0])
print('97th percentile:', percentiles[1])
print('99th percentile:', percentiles[2])
```

## Basic Outlier Detection

```python
PERCENTILE = 99
upper_threshold = np.percentile(y, PERCENTILE)
print(f'Upper threshold: {upper_threshold:.4f}')


# Get the y values of outliers
outliers_y = y[y>=upper_threshold]
# Get the x indices of outliers
outliers_x = x[np.argwhere(y>=upper_threshold)]
print(f'Outlier values: {outliers_y}')


plt.plot(x, y, c='blue', alpha=0.4, label='actual')
plt.scatter(outliers_x, outliers_y, c='red', alpha=0.9, label='outliers')
plt.legend()
```
